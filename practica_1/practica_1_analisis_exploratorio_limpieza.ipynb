```python
# Rutear la ubicación de las bases de datos

path_poblacion = r'.\enigh2020_est_poblacion_csv\poblacion.csv'
path_viviendas = r'.\enigh2020_est_viviendas_csv\viviendas.csv'
```

```python
# Cargar las bases de datos. En este caso, tomamos 2 bases de datos para unir campos de interés de cada una. 

poblacion = pd.read_csv(path_poblacion)
viviendas = pd.read_csv(path_viviendas)
```

```python
valores_parentesco = sorted(df_conjunto.parentesco.unique())
valores_nivelaprob = sorted(df_conjunto.nivelaprob.unique())
```

```python
etiquetas_parentesco = ['Jefe(a)', 'Esposo(a), compañero(a), cónyuge, pareja, marido, mujer, señor(a), consorte','Concubino(a)',
                       'Pareja del mismo sexo','Hijo(a), hijo(a) consanguíneo, hijo(a) reconocido',
                       'Hijo(a) adoptivo(a)','Hijastro(a), entenado(a)','Trabajador(a) doméstico(a)',
                       'Hijo(a) del(la) trabajador(a) doméstico(a)', 'Otro pariente del(la) trabajador(a) doméstico(a)',
                       'No tiene parentesco', 'Madre, padre', 'Padrastro, madrastra', 'Hermano(a)', 'Medio(a) hermano(a)','Abuelo(a)',
                       'Nieto(a)', 'Bisnieto(a)', 'Tataranieto(a)', 'Tío(a)', 'Sobrino(a)', 'Primo(a)', 'Suegro(a)',
                       'Nuera, yerno','Cuñado(a)', 'Concuño(a)', 'Ahijado(a)', 'Familiar, otro parentesco','Huésped, abonado(a), pensionista']
```

```python
etiquetas_nivelaprob = ['', 'Ninguno', 'Preescolar', 'Primaria', 'Secundaria', 'Preparatoria o bachillerato',
                       'Normal', 'Carrera técnica o comercial', 'Profesional', 'Maestría', 'Doctorado']
```

```python
df_conjunto.parentesco = df_conjunto.parentesco.replace(valores_parentesco, etiquetas_parentesco)
df_conjunto.nivelaprob = df_conjunto.nivelaprob.replace(valores_nivelaprob, etiquetas_nivelaprob)
```

```python
# Asignar etiquetas a valores(códigos númericos) de la base de datos

df_conjunto.sexo = df_conjunto.sexo.replace(to_replace = [1,2], value = ['Hombre', 'Mujer'])
df_conjunto.tipo_viv = df_conjunto.tipo_viv.replace(['1','2','3','4','5', '&'], ['Casa independiente', 'Departamento en edificio','Vivienda en vecindad', 'Vivienda en cuarto de azotea', 'Local no construido para habitación', 'No especificado'])
df_conjunto.mat_pared = df_conjunto.mat_pared.replace([1,2,3,4,5,6,7,8], ['Material de desecho', 'Lámina de cartón','Lámina de asbesto o metálica', 'Carrizo bambú o palma','Embarro o bajareque', 'Madera', 'Adobe', 'Tabique, ladrillo, block, piedra, cantera, cemento o concreto'])
df_conjunto.mat_pisos = df_conjunto.mat_pisos.replace([1,2,3], ['Tierra', 'Cemento o firme', 'Madera, mosaico u otro recubrimiento'])
df_conjunto.cocina = df_conjunto.cocina.replace([1,2], ['Sí', 'No'])
df_conjunto.cocina_dor = df_conjunto.cocina_dor.replace([1,2], ['Sí', 'No'])
df_conjunto.disp_agua = df_conjunto.disp_agua.replace([1,2,3,4,5,6,7], ['Agua entubada dentro de la vivienda', 'Agua entubada fuera de la vivienda, pero dentro del terreno', 'Agua entubada de llave pública (o hidrante)', 'Captadores de agua de lluvia', 'Agua entubada que acarrean de otra vivienda','Agua de pipa', 'Agua de un pozo, río, lago, arroyo u otra'])
df_conjunto.dotac_agua = df_conjunto.dotac_agua.replace([1,2,3,4,5], ['Diario', 'Cada tercer día', 'Dos veces por semana', 'Una vez por semana','De vez en cuando'])
df_conjunto.excusado = df_conjunto.excusado.replace([1,2], ['Sí', 'No'])
df_conjunto.sanit_agua = df_conjunto.sanit_agua.replace([1,2,3], ['Tiene descarga directa de agua', 'Le echan agua con cubeta', 'No se le puede echar agua'])
df_conjunto.drenaje = df_conjunto.drenaje.replace([1,2,3,4,5], ['La red pública', 'Una fosa séptica', 'Una tubería que va a dar a una barranca o grieta','Una tubería que va a dar a un río, lago o mar', 'No tiene drenaje'])
df_conjunto.disp_elect = df_conjunto.disp_elect.replace([1,2,3,4,5], ['Del servicio público', 'De una planta particular', 'De panel solar','De otra fuente', 'No tiene luz eléctrica'])
df_conjunto.tenencia = df_conjunto.tenencia.replace([1,2,3,4,5,6], ['Es rentada','Es prestada', 'Es propia pero la están pagando','Es propia', 'Está intestada o en litigio', 'Otra situación'])
df_conjunto.tam_loc = df_conjunto.tam_loc.replace([1,2,3,4], ['Localidades con 100 000 y más habitantes', 'Localidades con 15 000 a 99 999 habitantes','Localidades con 2 500 a 14 999 habitantes', 'Localidades con menos de 2 500 habitantes'])
df_conjunto.est_socio = df_conjunto.est_socio.replace([1,2,3,4], ['Bajo', 'Medio bajo', 'Medio alto', 'Alto'])
```

```python
df_conjunto['ubica_geo'] = df_conjunto['ubica_geo'].apply(lambda x: '{0:0>5}'.format(x))
```

```python
df_conjunto['cve_ent'] = df_conjunto['ubica_geo'].str.slice(start=0, stop=2)
```
